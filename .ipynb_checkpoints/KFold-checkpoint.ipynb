```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('./USA_Housing.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isnull.sum()
```

```python
data.isna.sum()
```

```python
house_data_x = data.drop('Price',axis=1)
```

```python
house_data_y = data['Price']
```

```python
house_data_x
```

```python
house_data_x.isnull()
```

```python
house_data_x.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
house_datax_scaled = scaler.fit_transform(house_data_x)
```

```python
house_datax_scaled
```

```python
house_datax_scaled = pd.DataFrame(house_datax_scaled,columns=house_data_x.columns)
```

```python
house_datax_scaled
```

```python
kf = KFold(n_splits = 5)
```

```python
from sklearn.metrics import r2_score
i=1
for train_index,test_index in kf.split(house_data_x,house_data_y):
    X_train, X_test = house_datax_scaled.iloc[train_index], house_datax_scaled.iloc[test_index]
    y_train, y_test = house_data_y.iloc[train_index], house_data_y.iloc[test_index]
    lreg = LinearRegression()
    lreg.fit(X_train,y_train)
    train_prediction = lreg.predict(X_train)
    test_prediction = lreg.predict(X_test)
    train_acc = r2_score(y_train,train_prediction)
    test_acc = r2_score(y_test,test_prediction)
    print(f'{i}th fold')
    print(f'train acc = {train_acc},test acc = {test_acc}')
    i=i+1

    
```
